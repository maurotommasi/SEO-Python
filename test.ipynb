```python
from SEO import SEO
```

```python
SEO_EN = SEO()
baseUrl = "edition.cnn.com"
onpage_seo_results, _ = SEO_EN.generate_seo_report(baseUrl)
```

```python
# Nested Analysis

nestedSEO = {}

for nestedLink in onpage_seo_results["Links"][0]:
    if baseUrl in nestedLink:
        nestedSEO[nestedLink], _ = SEO_EN.generate_seo_report(nestedLink, save=True, folder_path="edition.cnn.com")
```
